```python
# %load ../loaders/imports.py
import sys, os
import numpy as np
import matplotlib.pyplot as plt
import itertools
import time
import pdb

from sklearn.linear_model.coordinate_descent import _alpha_grid
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LassoCV, ElasticNetCV

from pyuoi.linear_model.cassolasso import PycLassoCV
from pyuoi.linear_model.casso_en import PycElasticNetCV, UoI_ElasticNet

sys.path.append('..')

from utils import gen_covariance, gen_beta2, gen_data, get_cov_list
from utils import selection_accuracy
from results_manager import init_results_container, calc_result
```

```python
np.linspace(0.1, 1, 10)
```

```python
# Initialize the data generating scenarios 

n_features = 100

# Block sizes
block_size = [10, 20]

# Block correlation
correlation = [0, 0.08891397, 0.15811388, 0.28117066, 0.5]

# Exponential length scales
L = [5, 10]

block_comb = list(itertools.product(correlation, block_size))
final_comb = []
final_comb.extend([(c[0], c[1], 1, 0) for c in block_comb])
final_comb.extend([(1, n_features, l, 1) for l in L])

cov_params = [{'correlation' : t[0], 'block_size' : t[1], 'L' : t[2], 't': t[3]} for t in final_comb]

sparsity = np.linspace(0.1, 1, 10)

betawidth = [np.inf, 0]

kappa = [100, 5]

iter_params = {'cov_params': cov_params, 'sparsity': sparsity, 'kappa': kappa, 'betawidth': betawidth}
const_params = {'n_features': n_features, 'n_samples': 400}

param_comb = list(itertools.product(*[iter_params[key] for key in iter_params.keys()]))

final_params = []
for i in range(len(param_comb)):
    arg_dict = const_params.copy()
    for j, key in enumerate(iter_params.keys()):
        arg_dict[key] = param_comb[i][j]
    arg_dict['index'] = i
    final_params.append(arg_dict)
    
final_params_rep = final_params * 10
```

```python
l1_ratios = [0.1, 0.5, 0.7, 0.9, 0.95]
```

```python
# Execute the 4 algorithms of interest
fields = ['FNR', 'FPR', 'sa', 'ee', 'median_ee', 'r2', 'MSE']
selection_methods = ['pyc_lasso', 'skl_lasso', 'pyc_en', 'skl_en']
results = init_results_container(selection_methods, fields, len(final_params_rep), 100)
skl_lasso_times = np.zeros(len(final_params_rep))
pyc_lasso_times = np.zeros(len(final_params_rep))

skl_en_times = np.zeros(len(final_params_rep))
pyc_en_times = np.zeros(len(final_params_rep))

for i, p in enumerate(final_params_rep):
    
    # Store the parameters
    for key, item in p.items():
        results[key] = item
    
    sigma = gen_covariance(p['n_features'], **p['cov_params'])
    beta = gen_beta2(p['n_features'], p['cov_params']['block_size'], sparsity = p['sparsity'], betawidth = p['betawidth'])
    X, X_test, y, y_test, ss = gen_data(p['n_samples'], p['n_features'], kappa = p['kappa'], 
                                        covariance = sigma, beta = beta)
    
    X = StandardScaler().fit_transform(X)
    y -= np.mean(y)
    
    alphas = _alpha_grid(X, y, n_alphas = 48)

    exp_results = []
    
    t0 = time.time()
    # Initalize sklearn lasso
    lasso1 = LassoCV(fit_intercept = False, alphas = alphas, cv = 5)
    lasso1.fit(X, y.ravel())
    skl_lasso_times[i] = time.time() - t0
    exp_results.append({})
    exp_results[0]['coefs'] = lasso1.coef_
    
    t0 = time.time()
    lasso2 = PycLassoCV(alphas, fit_intercept = False)
    lasso2.fit(X, y.ravel())
    pyc_lasso_times[i] = time.time() - t0    
    exp_results.append({})
    exp_results[1]['coefs'] = lasso2.coef_

    
    t0 = time.time()
    en1 = ElasticNetCV(l1_ratio = l1_ratios, n_alphas = 48, fit_intercept = False, cv = 5)
    en1.fit(X, y.ravel())
    skl_en_times[i] = time.time() - t0
    exp_results.append({})
    exp_results[2]['coefs'] = en1.coef_

    
    # Use Pycasso implementation of UoIElasticNet to generate the regularization parameters
    u = UoI_ElasticNet(n_lambda1 = 48, n_lambda2 = 5, fit_intercept= False)
    u.get_reg_params(X, y)

    l1 = u.lambda1[0] 
    l2 = u.lambda2
        
    t0 = time.time()
    en2 = PycElasticNetCV(l1, l2,fit_intercept = False, cv = 5)
    en2.fit(X, y.ravel())
    pyc_en_times[i] = time.time() - t0
    exp_results.append({})
    exp_results[3]['coefs'] = en2.coef_
        
    # perform calculations
    for j, selection_method in enumerate(selection_methods):
        for field in fields:
            results[selection_method][field][i] = calc_result(X, X_test, y, y_test, 
                                                                   beta.ravel(), field, 
                                                                   exp_results[j])
        
    if i % 10 == 0:
        print(i)
```

```python
import pandas as pd
final_params_rep = pd.DataFrame(final_params_rep)
```

```python
from misc import calc_avg_cov
```

```python
# Fix SNR, betawidth, and plot selection accuracy and estimation error
d = final_params_rep.loc[(final_params_rep['kappa'] == 5) & (final_params_rep['betawidth'] == np.inf)]

# Plot colormaps of sparsity vs. average corrrelation
sparsity = np.unique(d['sparsity'].values)

# track differences between the two solvers
lasso_sa_diff = np.zeros((10, 12))
en_sa_diff = np.zeros((10, 12))

lasso_ee_diff = np.zeros((10, 12))
en_ee_diff = np.zeros((10, 12))

for i, s in enumerate(sparsity):
    dd = d.loc[d['sparsity'] == s]
    
    avg_cov = np.zeros(dd.shape[0])
    
    # calculate average covariance
    for j in range(dd.shape[0]):
        
        indices = np.unique(dd['index'])
        for k, idx in enumerate(indices):
            
            ddd = dd.loc[dd['index'] == idx]
            for l in ddd.shape[0]:
                avg_cov[j] += calc_avg_cov(n_features, **ddd.iloc[l]['cov_params'])

                lasso_sa_diff[i, j] += results['skl_lasso']['sa'][ddd.iloc[l]['index']] - \
                                      results['pyc_lasso']['sa'][ddd.iloc[l]['index']]

                en_sa_diff[i, j] +=  results['skl_en']['sa'][ddd.iloc[l]['index']] - \
                                    results['pyc_en']['sa'][ddd.iloc[l]['index']]

                lasso_ee_diff[i, j] += (results['skl_lasso']['ee'][ddd.iloc[l]['index']] - \
                                      results['pyc_lasso']['ee'][ddd.iloc[l]['index']])\
                                      /results['skl_lasso']['ee'][ddd.iloc[l]['index']]

                en_ee_diff[i, j] +=  (results['skl_en']['ee'][ddd.iloc[l]['index']] - \
                                    results['pyc_en']['ee'][ddd.iloc[l]['index']])\
                                    /results['skl_en']['ee'][ddd.iloc[l]['index']]

    avg_cov *= 0.1
    lasso_sa_diff *= 0.1
    en_sa_diff *= 0.1
    lasso_ee_diff *= 0.1
    en_ee_diff *= 0.1
                                
fig, ax = plt.subplots(2, 2, figsize = (10, 10))

a1 = ax[0, 0].pcolormesh(lasso_sa_diff)
fig.colorbar(a1, ax = ax[0, 0])

a2 = ax[0, 1].pcolormesh(lasso_ee_diff)
fig.colorbar(a2, ax = ax[0, 1])

a3 = ax[1, 0].pcolormesh(en_sa_diff)
fig.colorbar(a3, ax = ax[1, 0])

a4 = ax[1, 1].pcolormesh(en_ee_diff)
fig.colorbar(a4, ax = ax[1, 1])
```
